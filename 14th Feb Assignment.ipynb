```python
# 1.Ans.

Multithreading in Python refers to the ability of a program to execute
multiple threads (subprocesses) simultaneously. 
Each thread runs independently and concurrently, allowing for parallel processing of tasks within a program.

Multithreading is used to improve the efficiency and performance of a
program by taking advantage of available computing resources, such as multiple CPU cores.
It allows a program to perform multiple tasks simultaneously and asynchronously, 
which can greatly improve the program's responsiveness and user experience.

The threading module in Python is used to handle threads.
```

```python
# 2.Ans.

The threading module in Python is used for creating and managing threads in a Python program.
Overall, the threading module in Python is a powerful tool for improving 
the performance, responsiveness, and efficiency of a program by allowing 
it to execute multiple tasks simultaneously.

activeCount(): The activeCount() function is a method of the threading module in Python. 
               It returns the number of Thread objects that are currently active in the Python interpreter. 
               The activeCount() function can be used to determine the current
               number of threads running in a program.

currentThread(): The currentThread() function is a method of the threading module in Python. 
                 It returns a reference to the current Thread object that is executing the current code. 
                 This function is useful for identifying the current thread of execution 
                 and can be used to access and manipulate the state of the current thread.

enumerate(): The enumerate() function is a built-in function in Python that is used to return 
             an iterator that contains tuples of (index, element) for a given iterable. 
             When used with threads, the enumerate() function can be used to get a list of all 
             the active Thread objects currently running in a program, along with their
             corresponding thread IDs. This function is useful for debugging and monitoring 
             the status of multiple threads in a program.
```

```python
# 3.Ans.

run(): The run() method is a method of the Thread class in Python. 
       It is called when a thread is started and is used to specify the 
       actions that the thread will perform. The run() method should be overridden 
       in a subclass of the Thread class to define the behavior of the thread.

start(): The start() method is a method of the Thread class in Python. 
         It is used to start the execution of a thread. The start() method creates 
         a new thread of execution and calls the run() method of the thread.

join(): The join() method is a method of the Thread class in Python. 
        It is used to wait for a thread to complete its execution. 
        The join() method blocks the current thread until the thread on which it is 
        called completes its execution. This method is useful for synchronizing the 
        execution of multiple threads.

isAlive(): The isAlive() method is a method of the Thread class in Python. 
           It is used to check whether a thread is currently executing or not. 
           The isAlive() method returns True if the thread is currently executing 
           and False otherwise. This method is useful for checking the status of a 
           thread and determining whether it has completed its execution or not.
```

```python
# 4.Ans.

import threading
def squares_number():
    for i in range(1,11):
        print(f"square of {i}:{i**2}")
        
def cubes_number():
    for i in range(1,11):
        print(f"cube of {i}:{i**3}")
        
if __name__ == "__main__":
    
    t1 = threading.Thread(target = squares_number)
    t2 = threading.Thread(target = cubes_number)
    
    t1.start()
    t2.start()
    
    t1.join()
    t2.join()
    
    print("Completed")
    
```

```python
# 5.Ans.

Advantages:

Increased performance: Multithreading can improve the performance of an
                       application by allowing multiple tasks to run in parallel.
Better resource utilization: Multithreading can make better use of system resources,
                            such as CPUs, by allowing multiple threads to run concurrently.
Improved responsiveness: Multithreading can make an application more responsive to
                         user input by allowing time-consuming tasks to be executed in the background.
Simplified programming: Multithreading can simplify programming by allowing complex 
                        applications to be broken down into smaller, 
                        more manageable parts that can be executed in parallel.
        
Disadvantages:

Increased complexity: Multithreading can increase the complexity of an
                      application, making it harder to develop, test, and maintain.
Synchronization issues: Multithreading can introduce synchronization issues,
                        such as race conditions, deadlocks, and livelocks, which 
                        can cause unpredictable behavior and bugs in the application.
Increased memory usage: Multithreading can increase the memory usage of an application,
                        as each thread requires its own stack and heap space.
GIL limitations: In Python, the Global Interpreter Lock (GIL) limits the execution of 
                 multiple threads within the same process. 
                 This means that multithreading may not provide a 
                 significant performance improvement in certain situations.
```

```python
# 6.Ans.

A deadlock occurs when two or more threads are waiting for each other to release
a resource that they need in order to proceed. In other words, each thread is holding
a resource that the other thread needs, and neither thread can proceed until the other 
releases its resource. This can result in a situation where the threads are stuck in a loop,
waiting for each other indefinitely.

A classic example of a deadlock is the "dining philosophers problem", in which a 
group of philosophers are seated at a table, each with a fork to their left and a
fork to their right. The philosophers need both forks to eat, but if they all grab 
their left fork at the same time, they will be deadlocked, as none of them will be able
to grab the fork on their right.

A race condition, on the other hand, occurs when two or more threads access
a shared resource in an unexpected or unpredictable order. This can result in
the incorrect or inconsistent behavior of the application.

For example, suppose two threads are trying to increment a shared counter variable. 
If they both read the current value of the variable, increment it, and then write the
new value back to the variable, they may end up overwriting each other's changes, resulting in a race condition.

To prevent deadlocks and race conditions, synchronization mechanisms such as 
locks, semaphores, and mutexes can be used to ensure that only one thread at a 
time can access a shared resource. Additionally, careful design and testing of concurrent
algorithms and data structures can help to avoid these issues.
```
